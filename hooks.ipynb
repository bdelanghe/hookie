```python
Scrabble['Length'] = Scrabble.Words.apply(lambda x: len(str(x)))
```

```python
uni_lengths = len(Scrabble.Length.unique())
```

```python
plt.hist(Scrabble['Length'],bins=uni_lengths)
plt.show()
```
